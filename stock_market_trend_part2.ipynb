```python
import tensorflow as tf
import pandas as pd 
import numpy as np
from tensorflow import keras
import matplotlib. pyplot as plt
%matplotlib inline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split


df1 = pd.read_csv("/content/drive/MyDrive/train.csv")
y = df1.Label
df1 = df1.drop(columns = ['Date ID' , "Label" , 'Actual Close Value'])
df1 , y
```

```python
scaler1 = MinMaxScaler()
df1= scaler1.fit_transform(df1)

df2 = pd.DataFrame(df1)
df2
```

```python
X = df1

from sklearn.model_selection import train_test_split
X_train , X_test ,y_train,  y_test = train_test_split(X , y , test_size = 0.1 , random_state = 0)


from keras import Sequential

model = Sequential([
    keras.layers.Dense(64 , input_shape = (21,) , activation = 'LeakyReLU' ),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(32 , activation = 'LeakyReLU' ),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(16 , activation = 'LeakyReLU' ),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(1 , activation = 'sigmoid' )

])

model.compile (
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']  
)

model.fit(X_train , y_train ,epochs = 200)
```

```python
model.evaluate(X_test , y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
pred_list = []
for items in y_pred:
  if items < 0.5:
    pred_list.append(0)
  else:
    pred_list.append(1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_list))
```
